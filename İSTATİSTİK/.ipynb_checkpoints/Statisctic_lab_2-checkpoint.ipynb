```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
tv_hours = [3,5,2,0,5,3,1,4,3,4]  # öğrencilerin TV izleme süreleri
GPA = [2.7,2.1,3.3,3.4,2.0,3.0,3.6,2.8,3.5,2.6]  # not ortalaması
```

```python
plt.scatter(tv_hours,GPA)
```

```python
plt.figure(figsize=(8,6))
ax = plt.axes()
ax.scatter(tv_hours, GPA)
#ax.plot(tv_hours,GPA)
ax.set_xlabel("TV Hours")
ax.set_ylabel("GPA")
ax.axis("tight")
plt.show()
```

```python
plt.figure(figsize=(8,6))
ax = plt.axes()
ax.scatter(tv_hours, GPA)
sns.regplot(tv_hours, GPA)
ax.set_xlabel('TV Hours')
ax.set_ylabel('GPA')
ax.axis('tight')
plt.show()
```

```python
x=np.array(tv_hours)
y=np.array(GPA)
```

```python
pay=len(x)*sum(x*y)-sum(x)*sum(y)
```

```python
payda=np.sqrt((len(x)*sum(x**2)-sum(x)**2)*(len(y)*sum(y**2)-sum(y)**2))
```

```python
pay
```

```python
payda
```

```python
pay/payda
```
