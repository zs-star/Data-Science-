```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
warnings.warn("this will not show")

plt.rcParams["figure.figsize"]=(10,6)
sns.set_style("whitegrid")
pd.set_option("display.float_format",lambda x:'%.3f'% x)
pd.options.display.max_rows=100
pd.options.display.max_columns=100
```

```python
Df=pd.read_csv("cleaned_AutoScout.csv")
```

```python
df=Df.copy()
```

```python
df.head(5)
```

### REORGANİZED VAT,TYPE,PAINT TYPE,EMISSION CLASS,EMISSION LABEL ACCORDING TO EFFECT ON PRICE

```python
df.groupby('vat')['price'].mean()##we can combine them their mean is not so different
```

```python
df.vat=df.vat.map(lambda x: 'deductible' if x=='Price negotiable' or x=='VAT deductible' else 'non-deductible')
```

```python
df.groupby(['Type','make_model'])['price'].mean()
```

```python
df.Type.value_counts()
```

```python
df.Type=df.Type.map(lambda x: "Used" if x=="Employee's car" else ( "New" if x=="Pre-registered" or x=="Demonstration" else x))
```

```python
df.groupby(['Body Color','make_model'])['price'].mean()
```

```python
df.groupby(['Paint Type','make_model'])[['price','Age']].mean()
```

```python
df["Paint Type"].value_counts()
```

```python
df["Paint Type"]=df["Paint Type"].map(lambda x: 'Non-metallic' if x=='Uni/basic' or x=='Perl effect' else x)
```

```python
df.groupby(['make_model','Emission Class'])[['price','Age']].mean()
```

```python
df[['Emission Label','Emission Class']].sample(5)
```

```python
df['Emission Label'].isnull().sum()##Drop
```

```python
df['Emission Class'].isnull().sum()
```

```python
df.drop('Emission Label',axis=1,inplace=True)
```

```python
df['Emission Class'].value_counts()
```

```python
df['Emission Class']=df['Emission Class'].map(lambda x: x.split(' ')[1][0],na_action='ignore')
```

```python
df[['Warranty','Age']].dropna()##There is no meaninful data
```

```python
df.drop('Warranty',axis=1,inplace=True)
```
