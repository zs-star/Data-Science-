```python
q1q3 = {}
for i in models:
    q1q3[i] = [np.percentile(df.price[df.make_model == i], 25)]
    q1q3[i].append(np.percentile(df.price[df.make_model == i], 75))
print(q1q3)
```

```python
iqr = {}
for i in q1q3:
    iqr[i] = q1q3[i][1] - q1q3[i][0]
print(iqr)    
    
```

```python
lower = {}
upper = {}
for i in q1q3:
    lower[i] = q1q3[i][0] - 1.5*iqr[i]
    upper[i] = q1q3[i][1] + 1.5*iqr[i]
print(lower)
print(upper) 
```

```python
for i in lower:
    print(i, len(df[(df.make_model == i) & (df.price < lower[i])]))
#Modellerdeki alt outlier sayıları
```

```python
for i in lower:
    df.drop(df[(df.make_model == i) & (df.price < lower[i])].index, inplace = True)
#Alt outlier olan değerleri drop ettim
```

```python
for i in lower:
    print(i, len(df[(df.make_model == i) & (df.price < lower[i])]))
#Kontrol ediyorum alt outlier değerler düşmüş mü diye
```

```python
for i in upper:
    print(i, len(df[(df.make_model == i) & (df.price > upper[i])]))
#Modellerdeki üst outlier sayıları
```

```python
for i in upper:
    df.drop(df[(df.make_model == i) & (df.price > upper[i])].index, inplace = True)
#Üst outlier olan değerleri drop ettim
```

```python
for i in upper:
    print(i, len(df[(df.make_model == i) & (df.price > upper[i])]))
#Kontrol ediyorum üst outlier değerler düşmüş mü diye
```
