## 2: Plot 12 most dangerous cities

```python
df.dropna(inplace=True)
```

```python
city=df.city.value_counts().sort_values(ascending=False).head(12)
city.index
city.values
```

```python
# with seaborn barplot
fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(city.index,city.values,ax=ax)
ax.set_xticklabels(city.index, rotation = 45)
ax.set_title('Most dangereous cities',color='blue',fontsize=15);
```

```python
# with seaborn countplot
fig, ax = plt.subplots(figsize=(8,6))
sns.countplot(x="city", data=df,order=df.city.value_counts().iloc[:12].index)
ax.set_xticklabels(city.index, rotation = 45);
ax.set_title('Most dangereous cities',color='blue',fontsize=15);
```

## 3: Plot 20 most dangerous states

```python
state=df.state.value_counts().sort_values(ascending=False).iloc[:20]
```

```python
# with seaborn barplot
fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(state.index,state.values,ax=ax)
ax.set_xticklabels(state.index, rotation = 45)
ax.set_title('Most dangereous states',color='blue',fontsize=15);
```

```python
# with seaborn barplot
```

```python
# with seaborn countplot
fig, ax = plt.subplots(figsize=(8,6))
sns.countplot(x="state", data=df,order=df.state.value_counts().iloc[:20].index)
ax.set_xticklabels(state.index, rotation = 45);
ax.set_title('Most dangereous states',color='blue',fontsize=15);
```
